Q1. You are a data analyst at an e-commerce company. Your task is to analyze sales data,
clean and process it, perform staƟsƟcal analysis, and generate visual insights to help the
business make informed decisions.
1. Use your roll number as the random seed to generate unique data for each student.
2. Follow the quesƟons step by step, as each part builds upon the previous one.
3. Submit the assignment as a Jupyter Notebook (.ipynb) with:
 Well-documented code and clear comments.
 Markdown explanaƟons for each step.
 Screenshots of the plots and outputs.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#  Step 1: Initialize Random Seed
roll_number = 102317008
np.random.seed(roll_number)

#  Step 2: Generate Sales Data (12 months, 4 product categories)
sales_data = np.random.randint(1000, 5001, size=(12, 4))

#  Step 3: Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun','Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, columns=categories, index=months)

#  Display first 5 rows
print(" First 5 Rows:")
print(df.head())

#  Summary statistics
print("\n Summary Statistics:")
print(df.describe())

#  Total Sales per Category
total_sales_by_category = df[categories].sum()
print("\n Total Sales per Category:")
print(total_sales_by_category)

#  Total Sales per Month
df['Total Sales'] = df[categories].sum(axis=1)
print("\n Total Sales per Month:")
print(df['Total Sales'])

#  Average Monthly Growth per Category
avg_growth = df[categories].pct_change().mean() * 100
print("\n Average Monthly Growth (%):")
print(avg_growth)

#  Growth Rate Column
df['Growth Rate'] = df['Total Sales'].pct_change() * 100
print("\n Data with Growth Rate:")
print(df[['Total Sales', 'Growth Rate']].head())

#  Apply Discount based on Roll Number
if roll_number % 2 == 0:
    df['Electronics (Discounted)'] = df['Electronics'] * 0.90
    print("\n 10% Discount applied to Electronics.")
else:
    df['Clothing (Discounted)'] = df['Clothing'] * 0.85
    print("\n 15% Discount applied to Clothing.")

#  Visualization 1: Line Plot for Monthly Sales Trends
plt.figure(figsize=(12, 6))
for cat in categories:
    plt.plot(df.index, df[cat], marker='o', label=cat)

plt.title("Monthly Sales Trends by Category")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

#  Visualization 2: Box Plot for Sales Distribution
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[categories], palette='pastel')
plt.title("Sales Distribution per Category")
plt.xlabel("Category")
plt.ylabel("Monthly Sales")
plt.grid(True, linestyle='--', axis='y')
plt.tight_layout()
plt.show()
```

Q.2 For the array: array = np.array([[1, -2, 3],[-4, 5, -6]])
i. Find element-wise absolute value
ii. Find the 25th, 50th, and 75th percenƟle of flaƩened array, for each column, for each
row.
iii. Mean, Median and Standard DeviaƟon of flaƩened array, of each column, and
each row

```python
import numpy as np

array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_array = np.abs(array)
print("Absolute value:\n", abs_array)

flat = array.flatten()

percentiles_flat = np.percentile(flat, [25, 50, 75])
print("\nPercentiles (Flattened): 25th, 50th, 75th:\n", percentiles_flat)

percentiles_col = np.percentile(array, [25, 50, 75], axis=0)
print("\nPercentiles (Each Column):\n", percentiles_col)

percentiles_row = np.percentile(array, [25, 50, 75], axis=1)
print("\nPercentiles (Each Row):\n", percentiles_row)

mean_flat = np.mean(flat)
median_flat = np.median(flat)
std_flat = np.std(flat)

print("\nFlattened - Mean:", mean_flat, "Median:", median_flat, "Std Dev:", std_flat)

mean_col = np.mean(array, axis=0)
median_col = np.median(array, axis=0)
std_col = np.std(array, axis=0)

print("\nPer Column - Mean:", mean_col)
print("Per Column - Median:", median_col)
print("Per Column - Std Dev:", std_col)

mean_row = np.mean(array, axis=1)
median_row = np.median(array, axis=1)
std_row = np.std(array, axis=1)

print("\nPer Row - Mean:", mean_row)
print("Per Row - Median:", median_row)
print("Per Row - Std Dev:", std_row)
```

Q.3 For the array: a = np.array([-1.8, -1.6, -0.5, 0.5,1.6, 1.8, 3.0]).
i. Find floor, ceiling and truncated value, rounded values

```python
import numpy as np

a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_vals = np.floor(a)
ceil_vals = np.ceil(a)
trunc_vals = np.trunc(a)
rounded_vals = np.round(a)

print("Original array:\n", a)
print("\nFloor values:\n", floor_vals)
print("\nCeiling values:\n", ceil_vals)
print("\nTruncated values:\n", trunc_vals)
print("\nRounded values:\n", rounded_vals)
```

Q.4 WAP to swap two elements in a list in Python Using a Temporary Variable.

```python
#sample listt
my_list = [10, 20, 30, 40, 50]

# Indices to swap
i = 1  # Index of first element (20)
j = 3  # Index of second element (40)

# Swap using a temporary variable
temp = my_list[i]
my_list[i] = my_list[j]
my_list[j] = temp

print("List after swapping:", my_list)
```

Q.5 Write a Python program to swap two elements in a set by converƟng it to a list first.

```python
# Original set
my_set = {10, 20, 30, 40, 50}

# Convert set to list
my_list = list(my_set)

# Indices to swap
i = 1
j = 3

# Swap elements using a temporary variable
temp = my_list[i]
my_list[i] = my_list[j]
my_list[j] = temp

# Convert back to set
swapped_set = set(my_list)

print("Original set:", my_set)
print("Set after swapping:", swapped_set)
```
