```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
nltk.download('punkt_tab')
text = "i love all kinds of food. i am fond of sea food especially prawns. i like pizza too. the crust and cheese just draw me towards them and the pizza sauce is just chef's kiss. also i love garlic bread."
text.lower()
tokens = word_tokenize(text)  # word tokenization
cleaned_tokens = [token for token in tokens if token.isalpha()]
print(cleaned_tokens)
```

```python
tokens2 = sent_tokenize(text)   # sentence tokenization
print(tokens2)
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python

filtered_list = []
stop_words = set(stopwords.words('english'))
for word in cleaned_tokens:
  if word not in stop_words:
    filtered_list.append(word)
print(filtered_list)
```

```python
from nltk.probability import FreqDist
freq_dist = FreqDist(filtered_list)
print('Word frequency distribution: ')
for word, freq in freq_dist.items():
  print(f"{word}: {freq}")

freq_dist.plot(title = 'Top words excluding stop words')
```

```python

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(word) for word in filtered_list]
print(stemmed_words)

```

```python
from nltk.stem.lancaster import LancasterStemmer
st = LancasterStemmer()
stemmed_words2 = [st.stem(word) for word in filtered_list]
print(stemmed_words2)
```

```python

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_list]
print(lemmatized_words)
```

```python
for a, b in zip(stemmed_words, lemmatized_words):
  print("Stemmed: {a}, Lemmatized: {b}")
```

```python
import re
long_words = re.findall(r'\b\w{6,}\b', text)
print(long_words)
```

```python
numbers = re.findall(r'\b\d+\b', text)
print(numbers)
```

```python
capitalized_words = re.findall(r'\b[A-Z][a-z]*\b', text)
print(capitalized_words)  # ['Alice', 'Mathematics', 'Bob', 'Catherine']
```

```python
pattern = r"\d+\.\d+|\w+(?:-\w+)+|\w+(?:'\w+)?"
custom_tokens = re.findall(pattern, text)
print(custom_tokens)
```

```python
text2 = 'contact me at www.hahahaha.com and the mail is hahamo@gmail.com, along with the phone number: 1234567890'
text2 = re.sub(r'\b[\w.-]+@[\w.-]+\.\w+\b', '', text2)
text2 = re.sub(r'https?://\S+|www\.\S+', '', text2)
text2 = re.sub(r'\b(?:\+?\d{1,3}[ -]?)?(?:\d{3}[- ]?\d{3}[- ]?\d{4})\b', '', text2)
print(text2)
```
