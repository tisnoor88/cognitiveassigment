Q1. Ask the user to enter a value (e.g., M) for a mathematical function.
 Generate x values from -10 to 10 using np.linspace().
 Compute y values for:
o y= M⋅x
2
o y=M⋅sin(x)
 Plot both functions in a single figure:
o Use different colors and line styles.
o Add legend, grid, and title.

```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, label='y = M * x^2', color='blue', linestyle='-')
plt.plot(x, y2, label='y = M * sin(x)', color='red', linestyle='--')
plt.grid(True)
plt.legend()
plt.title(f"Plots of y = {M} * x^2 and y = {M} * sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Q.2 Create a dataset of five subjects and scores.
 Convert it into a Pandas DataFrame.
 Plot the scores using a Seaborn bar plot with:
o Different colors for each bar.
o Annotations on top of each bar.
o Title, axis labels, and grid.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subjects = ['Math', 'Science', 'English', 'History', 'Art']
scores = [88, 92, 79, 85, 90]

df = pd.DataFrame({'Subject': subjects, 'Score': scores})

plt.figure(figsize=(8, 6))
bar_plot = sns.barplot(x='Subject', y='Score', data=df, palette='Set2')

for i, score in enumerate(df['Score']):
    plt.text(i, score + 1, str(score), ha='center', va='bottom', fontweight='bold')

plt.title("Scores by Subject")
plt.xlabel("Subject")
plt.ylabel("Score")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, max(scores) + 10)
plt.tight_layout()
plt.show()
```

Q3. Write a Python script to select your roll number as seed for NumPy and
then generate a dataset of 50 values using np.random.randn().
 Create a 2x2 subplot layout displaying:
o Line plot showing cumulative sum.
o Scatter plot with random noise.
 Customize with titles, axis labels, and grids.

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = 102317008
np.random.seed(roll_number)

data = np.random.randn(50)
cumsum_data = np.cumsum(data)
noise = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(cumsum_data, color='blue')
axs[0, 0].set_title("Cumulative Sum Line Plot")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Cumulative Sum")
axs[0, 0].grid(True)

axs[0, 1].scatter(data, noise, color='green', alpha=0.7)
axs[0, 1].set_title("Scatter Plot with Random Noise")
axs[0, 1].set_xlabel("Original Data")
axs[0, 1].set_ylabel("Noise")
axs[0, 1].grid(True)

axs[1, 0].axis('off')
axs[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

Q.4 Download Data-set from the below link
https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv
Apply ‘seaborn’ library to do the following.
1. Read Total profit of all months and show it using a line plot.
2. Read all product sales data and show it using a multiline plot.
3. Plot bar chart for all the features/attributes.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset directly from GitHub (raw link)
url = 'https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv'
df = pd.read_csv(url)

# 1. Line plot: Total Profit
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='month_number', y='total_profit', marker='o', color='green')
plt.title('Company Total Profit per Month')
plt.xlabel('Month')
plt.ylabel('Total Profit')
plt.grid(True)
plt.tight_layout()
plt.show()

# 2. Multiline plot: Product sales
plt.figure(figsize=(12, 7))
product_columns = df.columns[1:-1]

for col in product_columns:
    sns.lineplot(data=df, x='month_number', y=col, label=col)

plt.title('Monthly Product Sales')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Bar chart: All attributes
df.set_index('month_number').plot(kind='bar', figsize=(14, 8))
plt.title('Monthly Sales and Profit Overview')
plt.xlabel('Month')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
